```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
sns.barplot(data=df)
plt.xticks(rotation=90)
plt.title("items distribution")
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
data=pd.read_csv('Market_Basket_Optimisation.csv')
data
```

```python
data.isnull().sum()
```

```python
records = []
for i in range(0, len(data)):
    records.append([str(data.values[i,j]) for j in range(0, 20)])
    records[i] = [x for x in records[i] if x != 'nan']
```

```python
records
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(records).transform(records)    #Apply one-hot-encoding on our dataset
data=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
data
```

```python

sns.barplot(data=data)
plt.xticks(rotation=90)
plt.title("items distribution")
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(data, min_support=0.05, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
